Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
import plotly.express as px
```

```python
scatter_plot=px.scatter(titanic,x="age",y="fare")
scatter_plot.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
box_plot=px.box(tips,x="day",y="total_bill")
box_plot.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
histogram=px.histogram(tips,x="sex",y="tip",pattern_shape="smoker",color="day")
histogram.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
iris_data=px.data.iris()
iris_data.head()
```

```python
scatter_matrix=px.scatter_matrix(iris_data,dimensions=["sepal_length","sepal_width","petal_length","petal_width"],color="species")
scatter_matrix.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

Ans 5. Displot is also known as distribution plot which is a graphical representation that is  combination of histogram and kernel density estimate. histogram provides a binned  representation of data's frequency and kernel density estimation provides a smooth representation  of probabiltiy density function.


```python
import numpy as np
```

```python
data=np.random.randn(1000)
```

```python
displot=px.histogram(data,nbins=30)
displot.show()
```
